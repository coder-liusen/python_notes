python内建的文件操作函数是open()，它返回文件流对象，而不是文件本身。
另外，numpy、pandas等库也有导入数据文件的方法。

### 文件操作的基本用法 

```python
# (1)打开文件：open()
# 函数原型：f = open(file,mode)
f = open('data/myfile.txt','r',encoding = 'utf-8')
# open()函数的返回值f是一个文件流对象
```

```python
# (1)关闭文件：close()
f = open('data/myfile.txt','r',encoding = 'utf-8')
f.close()
```

```python
# (2)读取指定字节数的文本（如不给参数则读取所有）：read()

f = open('data/myfile.txt','r',encoding = 'utf-8')
text_all = f.read()
print('all cnotent:\n',text_all)

f = open('data/myfile.txt','r',encoding = 'utf-8')
text_5 = f.read(5)
print('head 5 content:\n',text_5)
```

```python
# (3)读取一行文本：readline()
f = open('data/myfile.txt','r',encoding = 'utf-8')
print(f.readline())
print(f.readline())
print(f.readline())
```

```python
# (4)读取所有文本，并分行保存成一个字符串列表：readlines()
f = open('data/myfile.txt','r',encoding = 'utf-8')
print(f.readlines())
```

```python
# (5)向文件中写入字符串：write()
f = open('data/myfile2.txt','w',encoding = 'utf-8')
num1 = f.write('haha')
print(num1)
num2 = f.write('nihao')
print(num2)
f.close()
```

```python
# (6)将字符串列表中的所有字符串依次写入文件：writelines()
f = open('data/myfile3.txt','w',encoding = 'utf-8')
lst = ['add1','add2','add3']
f.writelines(lst)
f.close()
```

注意，python文件操作中，二进制文件可以不使用缓冲，但是文本文件必须使用缓冲。

### 文件打开模式

### open()用作上下文管理器

```python
with open('data/myfile.txt','r',encoding='utf-8') as f:
    print(f.read())
    
# with用于创建一个临时的运行环境，运行环境中的代码执行完后自动安全退出环境。
#在with管辖范围内的代码中，对f的操作过程中可能会产生的异常和运行完后资源的释放，都由上下文管理器open来处理。
# https://www.cnblogs.com/dingjiaoyang/p/11004065.html
```
