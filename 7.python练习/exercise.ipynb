```python
## 1.输入年月日，给出该日期是该年中的第几天

def cd(year,month,day):

    md = [31,28,31,30,31,30,31,31,30,31,30,31] # 非闰年各月天数
    md_r = [31,29,31,30,31,30,31,31,30,31,30,31] # 闰年各月份天数

    if(year % 4): # 非闰年
        
        if((year < 0) or (month not in range(1,13)) or (day not in range(1,md[month-1]+1))):
           return '输入错误!'
        
        d_th = sum(md[:month-1]) + day

    else: # 闰年
        if((year < 0) or (month not in range(1,13)) or (day not in range(1,md_r[month-1]+1))):
           return '输入错误!'

        d_th = sum(md_r[:month-1]) + day

        return d_th

## test
year = 2020
month = 2
day = 29
print(cd(year,month,day))
```

```python
## 2.统计一个范围内的平方数个数（用filter函数实现）

def ifsquare(a):

    import math
    sa = math.sqrt(a)
    return round(sa) == sa


def countsqu(a,b):

    if(type(a) != int):

        print("区间左非整数！")
        return None

    if(type(b) != int):

        print("区间右非整数！")
        return None

    if(a > b):
        print("区间左大于区间右！")
        return None

    return len(list(filter(ifsquare,list(range(a,b)))))


## test
a = 100
b = 168
print(countsqu(a,b))
```

```python
## 3.统计字符串中每个字符出现的次数

def countstr_1(st): #方法一

    dic_st = {}
    for c in st:
        dic_st[c] = dic_st.get(c,0) + 1 # 字典的get方法，若获取不到值，可以指定一个返回值

    return dic_st

def countstr_2(st): #方法二
    
    set_st = set(st)
    dic_st = {}.fromkeys(set_st,0)
    for c in st:
        dic_st[c]+=1

    return dic_st

def countstr_3(st):
    from collections import Counter
    return dict(Counter(list(st)))

## test
s = 'ghtaagbnbb'
print('方法一：',countstr_1(s))
print('方法二：',countstr_2(s))
print('方法三：',countstr_3(s))
```

```python
## 4.练习装饰器

# (1) 打印函数名
import functools

def printname(fn):

    @functools.wraps(fn)
    def wrapper(*args,**kws):
        print('函数名字是： %s' %(fn.__name__))
        return fn(* args,**kws)

    return wrapper


@printname
def fn(a,b,AA,BB):

    print(a)
    print(b)
    print(AA)
    print(BB)

    
## test
a = 1
b = 2
AA = "aa"
BB = "bb"

fn(b = 7.8,a = 3.6,BB = "66",AA = "77")

# (2) 打印函数名和函数运行时间
import time,functools

def printtime(fn):

    @functools.wraps(fn)
    def wrapper(*args,**kw):
        time0 = time.time()
        ret = fn(*args,**kw)
        time1 = time.time()
        print('%s 函数执行时间为 %s ms' %(fn.__name__,time1 - time0))

        return ret

    return wrapper

## test
@printtime
def funct(a):
    for i in range(1,10000):
        a *= 1/i

    return a

funct(8)
```

```python
# 5.删除列表中重复元素
def deldupli_sorted(lst): # 排序

    return list(set(lst))


def deldupli_unsorted_1(lst): # 不排序，方法一

    lst_rec = []
    for i in range(len(lst)):

        if lst[i] not in lst_rec:
            lst_rec.append(lst[i])

    return lst_rec

def deldupli_unsorted_2(lst): # 不排序,方法二（列表推导式）

    return [lst[i] for i in range(len(lst)) if lst[i] not in lst[:i]]

## test
lst = [4,3,5,7,2,3,6,5,2,2,7]

for item in deldupli_sorted(lst):
    print(item,end = " ")
print('\n')
for item in deldupli_unsorted_1(lst):
    print(item,end = " ")
print('\n')
for item in deldupli_unsorted_2(lst):
    print(item,end = " ")
```

```python
# 6.输入日期，输出该日期值什么班
# 值班规律：主班、中午班、夜班、下夜班(循环)
# 起始时间：2020.9.11 周五 主班

from datetime import date,datetime

def duty(str_date):

    try:
        date_input = datetime.strptime(str_date,'%Y-%m-%d').date()
    except:
        print('输入错误！')
        return False

    date_start = date(2020,9,11)
    if(date_input < date_start):
        print('输入日期必须在2020年9月11日之后！')
        return False
    
    print('这一天是星期%d' %date_input.isoweekday())

    d = (date_input - date_start).days % 4
    dic_duty = {0:'这一天是主班',1:'这一天是中午班',2:'这一天是夜班',3:'这一天是下夜班'}
    print(dic_duty.get(d,'计算出错！'))
    
    return True

## test
while(1):
    str_date = input('请按此格式输入日期：year-month-day（若要结束程序，请输入quit）：')
    if(str_date == 'quit'):
        break
    else:
        duty(str_date)
```

```python
## 7.含重复元素的有序数列二分查找

def find_min(a,t):

    tem = t + 1
    left = 0
    right = len(a) - 1

    while(True):
        tem = a[(left + right)//2]
        if(tem < t):
            left = (left + right)//2
        elif(tem >= t):
            right = (left + right)//2

        if(left == right - 1):
            break

    print('最小索引 :',right)

def find_max(a,t):

    tem = t + 1
    left = 0
    right = len(a) - 1

    while(True):
        tem = a[(left + right)//2]
        if(tem <= t):
            left = (left + right)//2
        elif(tem > t):
            right = (left + right)//2

        if(left == right - 1):
            break

    print('最大索引 :',left)
    
## test
a = [1,5,6,7,7,7,8,10,10,10,21,22,23,23,45]

find_min(a,23)
find_max(a,23)
```

```python
## 8.取中位数

def median(lst):

    if(not lst):
        print("Error! 输入列表不能空！")
        return None
    
    lst.sort()
    med = lst[len(lst)//2] if len(lst)%2 \
                           else (lst[len(lst)//2-1]+lst[len(lst)//2])/2

    return med

## test
import random
lst = list(range(10))
random.shuffle(lst)
print("返回值为：",median(lst))
```

```python
## 15.用 __new__实现单例模式
# （实例化一个类时，必调用__new__函数，所以实现一个单例类，
# 只需要让__new__函数无论多少次实例化，都只返回一个实例对象即可）

class singleton(object):

    __instance = None
    name = None
    age = None

    def __new__(cls,age,name):

        if not cls.__instance:
            cls.__instance = object.__new__(cls)

        return cls.__instance

    def __init(self,age,name):

        self.age = age
        self.name = name
```

```python
## 17.随机生成电话号码并写入文件（练习字符串、随机函数、文件操作）
import random

n=20
lst = []
a = list(range(10))

## 随机生成北京地区固定电话号码
for i in range(n):
    random.shuffle(a)
    lst.append("010-"+"".join(str(a[:8]).strip('[').strip(']').split(', ')))

## 写入文件
with open('file.txt','w') as f:
    for i in range(len(lst)):
        f.write(lst[i]+'\n')
```

```python
## 18.去除首尾空格
def trim(s):

    while s[:1] == " ":
        s = s[1:]
    while s[-1:] == " ":
        s = s[:-1]

    return s
```

```python
## 19.计算一个字符串中大写字母的数量（用filter实现,或用列表推导式实现）
def uppernum_1(s):
    
    return len(list(filter(lambda c:True if c.isupper() else False,s))) # filter函数

def uppernum_2(s):
    
    return len([c for c in s if c.isupper()]) #列表推导式

# test
s = 'hsdLDJFlldKlkdfjLDjlfdjflDFJlkdfjLKD'
print(uppernum_1(s))
print(uppernum_2(s))
```
